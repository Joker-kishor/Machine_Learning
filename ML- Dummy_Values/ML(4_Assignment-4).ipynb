```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df.head(3)
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
final = pd.concat([df ,dummies],axis='columns')
final
```

```python
result = final.drop(['town','west windsor'],axis='columns')
result.head(3)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = result.drop(['price'],axis='columns')
X.head(3)
```

```python
y = result['price']
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(X,y)
```
